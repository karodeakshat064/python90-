Lists and Lists comprehension

```python
# list creation
countries = ['india','us','england']
# print whole list
print(countries)
# accessing list elements
print("value at index 0:",countries[0])
# assigning new value at index
countries[0]='canada'
print("new value",countries[0])
```

Iterating list

```python
hybridlists=[' My name is anya jain',
             ' im in 3rd year']
print(hybridlists)
# we can access the elements of list using loop
for item in hybridlists:
  print(item)
```

```python
# print the index number with value
countries=['india','austrailia','usa']
for index,country in enumerate(countries):
  place=str(index)
  print("index: "+place+" value : "+country)
# find index of
index=countries.index('india')
print('index of india is ',index)
```

```python
# remove the last country from list by index
del countries[-1]
print(countries)
```

For loop


```python
# normal for loop
l=[1,2,3,4,4]
for i in l:
  print(i)
```

```python
# using list comprehension construct a list from
# if a square is greater than 50
l=[2,4,6,8,10,114]
[i**2 for i in l if i**2>50]
```

```python
# dictionary is consisted of behicles
# construct a list of names of vehicles
# the key names all upper case
dict={"sedan":1500,"SUV":2000,"pickup":2500}
d=[i.lower() for i in dict if dict[i]<5000]
print(d)
```

```python
# if else statement if value is greater than 5 print high else print low
l=[1,2,3,4,5,6,7,8,9]
l2=['high' if i>5 else 'low' for i in l]
print(l2)
```

```python
# if x to the power of 2 is less than or equal to 10 execute 27^x for values in range 1 to 20
[27*x for x in range(1,21) if x**2<=10]
```

```python
# even number in range
list=[i for i in range(51) if i%2==0]
print(list)
```

```python
# numbers that can be divided by 2 and 5
[x for x in range(51) if x%2==0 and x%5==0]
```

```python
# print table of 5 from 1 to 30
l=[5*x for x in range(1,31)]
print(l)
```

```python
# count the number of vowels
string =" i love this csit department"
j=[]
for i in string:
  if(i in['a','e','i','o','u']):
    j.append(i)
print("Number of vowels:", len(j))
```

```python
# given a list of numbers,remove all odd numbers
numbers=[3,5,45,97,32,22,10,19,39,43]
result=[]
for number in numbers:
  if number%2==0:
    result.append(number)
print(result)
```

```python
'''
get only numbers in sentence
'''
sentence='in 1984 there were 13 instances of a something 1000 doesnt matter'
words =  sentence.split()
result=[number for number in words if not number.isalpha()]
print(result)
```

```python
'''
find the common numbers in two lists(without using a tuple or set)
'''
list_a =[1,2,3,4]
list_b=[2,3,4,5]
common=[a for a in list_a if a in list_b]
print(common)
```
