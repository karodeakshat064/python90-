[1,2,3,4,5] → These are your x-values.

[1,4,9,16,25] → These are your y-values (squares of x-values).

'r^-' → This is a format string that styles the plot:

'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

```python
plt.grid(True)
plt.title('squares of the numbers')
plt.xlabel('The Numbers', fontsize=18, color='red')
plt.ylabel('The squares',fontsize=18, color='red')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r^--', linewidth=3.0)
plt.show()
```

### 3D Surface Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Assuming x and y are available as lists from previous cells
# Convert x and y to numpy arrays for meshgrid
X = np.array(x)
Y = np.array(y)

# Create a meshgrid
X_mesh, Y_mesh = np.meshgrid(X, Y)

# Define a function for Z values (e.g., Z = X^2 + Y^2)
Z_surface = X_mesh**2 + Y_mesh**2

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X_mesh, Y_mesh, Z_surface, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (X^2 + Y^2)')
ax.set_title('3D Surface Plot')

plt.show()
```

### 3rd 3D Surface Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Re-using X_mesh and Y_mesh from previous cell

# Define a different function for Z values (e.g., a more complex surface)
Z_surface_3rd = np.sin(np.sqrt(X_mesh**2 + Y_mesh**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X_mesh, Y_mesh, Z_surface_3rd, cmap='plasma') # Using a different colormap

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (sin(sqrt(X^2 + Y^2)))')
ax.set_title('3rd 3D Surface Plot')

plt.savefig('3rd_surface_plot.png') # Save the plot as a PNG image
plt.show()
```

### 3D Parametric Curve (Helix)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the parameter t
t = np.linspace(-2 * np.pi, 2 * np.pi, 100)

# Define x, y, z as functions of t for a helix
x_curve = np.sin(t)
y_curve = np.cos(t)
z_curve = t

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the parametric curve
ax.plot(x_curve, y_curve, z_curve, color='purple', linewidth=3, label='3D Helix Curve') # Added label

ax.set_xlabel('X-axis (sin(t))')
ax.set_ylabel('Y-axis (cos(t))')
ax.set_zlabel('Z-axis (t)')
ax.set_title('3D Parametric Curve (Helix)')

ax.legend() # Added legend
plt.savefig('parametric_curve.png') # Save the plot as a PNG image
plt.show()
```
